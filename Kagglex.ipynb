```python
fig = px.histogram(train,
             x='model_year',
             marginal='box',
             nbins=50,
             title='Car Distribution by Model Year')
fig.update_layout(bargap=0.1)
fig.show()
```

```python
fig = px.histogram(train,
             x='milage',
             marginal='box',
             nbins=41,
             title='Car Distribution by Mileage')
fig.update_layout(bargap=0.1)
fig.show()
```
